Create a `Book` class with attributes `title`, `author`, and `pages`.<br>
Add a class attribute `library_name` with the value "BNF".<br>
Add a method `display_info()` that prints the book's information.<br>
Add a method `is_long_book()` that returns `True` if the book has more than 300 pages.

```python
class Book:
    library_name = "BNF"  # shared class attribute

    def __init__(self, title, author, pages):
        self.title = title   # instance attribute
        self.author = author
        self.pages = pages

    def display_info(self):
        # displays book information
        print(f"'{self.title}' by {self.author} ({self.library_name})")

    def is_long_book(self):
        return self.pages > 300
```

Instantiate two different books: `Python Basics` and `Advanced Python` (other attributes of your choice) and display their information. Test the `is_long_book()` method.

```python
book1 = Book("Python Basics", "Alice", 250)
book2 = Book("Advanced Python", "Bob", 400)

book1.display_info()   # 'Python Basics' by Alice (Central Library)
book2.display_info()   # 'Advanced Python' by Bob (Central Library)

print(book1.is_long_book())  # False
print(book2.is_long_book())  # True
```

Add a dynamic attribute 'genre' to the first book and display it.

```python
book1.genre = "Programming"
print(book1.genre)  # Programming
```

Create a subclass `EBook` that inherits from `Book`.<br>
Add an additional attribute `file_size` (in MB).<br>
Override `display_info()` to also show the file size.

```python
class EBook(Book):
    def __init__(self, title, author, pages, file_size):
        super().__init__(title, author, pages)
        self.file_size = file_size

    def display_info(self):
        print(f"'{self.title}' by {self.author} ({self.library_name}), File size: {self.file_size} MB")
```

Instantiate an `EBook` and test `display_info()` and `is_long_book()`.

```python
ebook1 = EBook("Python E-Book", "Carol", 320, 5)
ebook1.display_info()  # 'Python E-Book' by Carol (Central Library), File size: 5 MB
print(ebook1.is_long_book())  # True
```
