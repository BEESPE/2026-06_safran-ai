## <font color="green">1- Days of the week </font>

Create a list `week` containing the 7 days of the week.

- From this list, how can you retrieve, on the one hand, the first 5 days of the week, and on the other hand, the weekend days, using positive indexing?
- Obtain the same result using negative indexing.
- Access the last day of the week.
- Reverse the days of the week in a single command.

```python
# Votre code
```

### <font color="red"> Solution </font>

```python
week = [
    "monday",
    "tuesday",
    "wednesday",
    "thursday",
    "friday",
    "saturday",
    "sunday",
]

print(week[:5])
print(week[5:])

print(week[:-2])
print(week[-2:])

print(week[6])
print(week[-1])

print(week[::-1])
```

## <font color="green">2- Seasons </font>

Create 4 lists `winter`, `spring`, `summer`, and `autumn` containing the corresponding months of each season. Then, create a list `seasons` containing the lists `winter`, `spring`, `summer`, and `autumn`. Predict what the following instructions will return, then check the result in the interpreter.

- `seasons[2]`
- `seasons[1][0]`
- `seasons[1:2]`
- `seasons[:][1]`

```python
# Votre code
```

### <font color="red"> Solution </font>

```python
winter = ["december", "january", "february"]
spring = ["march", "april", "may"]
summer = ["june", "july", "august"]
autumn = ["september", "october", "november"]

seasons = [winter, spring, summer, autumn]

print(seasons[2])
print(seasons[1][0])
print(seasons[1:2])
print(seasons[:][1])
```
