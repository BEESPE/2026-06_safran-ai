Write a program that, given two variables — one representing a sequence as a string, and the other an integer $n$ — computes all subsequences of $n$ letters that appear in the sequence and their occurrences, in the form of a dictionary, and displays it.

For example, for the sequence `le chat ronronne` and $n = 2$, the displayed dictionary would be:

```python
{'le': 1, 'e ': 1, ' c': 1, 'ch': 1, 'ha': 1, 'at': 1, 't ': 1, ' r': 1, 'ro': 2, 'on': 2, 'nr': 1, 'nn': 1, 'ne': 1}
```

### <font color="red"> Solution </font>

````python
sequence = "le chat ronronne"
n = 2
resultat = {}
for i in range(len(sequence) - n + 1):
    n_gramme = sequence[i:i + n]
    if n_gramme not in resultat:
        resultat[n_gramme] = 1
    else:
        resultat[n_gramme] += 1
print(resultat)
````

````python
def n_grammes(sequence, n=2):
    result = {}
    for i in range(len(sequence) - n + 1):
        n_gramme = sequence[i:i + n]
        if n_gramme not in result:
            result[n_gramme] = 1
        else:
            result[n_gramme] += 1
    return result


print(n_grammes("le chat ronronne", 2))
````

## <font color="green">2- Which person should you choose? </font>

You want to hire someone for a mission involving C++ and Python. You have the following information (in the form of a dictionary with a name as key and skills as value).

```python
skills_people = {
    'Alice': ['SAS', 'Python', 'R'],
    'Bob': ['HTML', 'JavaScript', 'CSS'],
    'Laura': ['C++', 'ASM', 'Python', 'CUDA']
}
mission = ['C++', 'Python']
```

Write a program that prints the names of qualified candidates. The solution must remain valid if the `competences` and `mission` variables change.

Optional: convert the mission list into a `frozenset`.

````python
skills_people = {'Alice': ['SAS', 'Python', 'R'],
               'Bob': ['HTML', 'JavaScript', 'CSS'],
               'Laura': ['C++', 'ASM', 'Python', 'CUDA']}
mission = ['C++', 'Python']

# Votre code
````

### <font color="red"> Solution </font>

````python
for person_name, person_skills in skills_people.items():
    good_fit = True

    for mission_skill in mission:
        if mission_skill not in person_skills:
            good_fit = False
    
    if good_fit:
        print(person_name)
````

````python
mission_set = frozenset(mission)
print(
    [
        nom
        for nom, person_skills in skills_people.items()
        if mission_set.issubset(person_skills)
    ]
)
````

````python
def filter_employees(skills_people, mission):
    mission_set = frozenset(mission)
    return [nom
            for nom, person_skills in skills_people.items()
            if mission_set.issubset(person_skills)]

filter_employees(skills_people, mission)
````

## <font color="green">3- Set manipulation </font>

Given the following sets describing characteristics of programming languages (you may consult the Python documentation on sets during the exercise):

- Find the characteristics common to JavaScript and Python
- Find the characteristics exclusive to Java compared to Python
- What is the complete set of characteristics?
- Which characteristics are exclusive to Python compared to all other languages?

````python
js = {"dynamic typing", "interpreted", "native to the web", "prototype-oriented"}
python = {"dynamic typing", "interpreted", "scientific computing", "object-oriented"}
java = {"compiled", "object-oriented", "static typing"}

languages = [js, python, java]
````

````python
# Votre code
````

### <font color="red"> Solution </font>

````python
print(
    "Features shared by JS and Python:",
    js & python
)

print(
    "Features exclusive to Java compared to Python:",
    java - python
)

print(
    "Complete set of features:",
    set().union(*languages)
)

print(
    "Features exclusive to Python compared to the others:",
    python.difference(js, java)
)
````
