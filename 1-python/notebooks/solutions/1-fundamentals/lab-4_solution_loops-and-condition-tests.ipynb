## <font color="green">2- Complementary sequence of a DNA strand</font>

This list represents a DNA sequence:

```python
["A", "C", "G", "T", "T", "A", "G", "C", "T", "A", "A", "C", "G"]
```

Create a script that transforms this sequence into its complementary sequence.

The complementary sequence is obtained by replacing `A` with `T`, `T` with `A`, `C` with `G`, and `G` with `C`.

````python
# Votre code
````

### <font color="red"> Solution </font>

````python
dna_strand = ["A", "C", "G", "T", "T", "A", "G", "C", "T", "A", "A", "C", "G"]

comp_dna_strand = []
for nitrogenous_base in dna_strand:
    if nitrogenous_base == "A":
        comp_dna_strand.append("T")
    elif nitrogenous_base == "T":
        comp_dna_strand.append("A")
    elif nitrogenous_base == "C":
        comp_dna_strand.append("G")
    else:
        comp_dna_strand.append("C")

print(comp_dna_strand)
````

## <font color="green">4- Amino acid frequency</font>

The list below represents an amino acid sequence:

```python
["A", "R", "A", "W", "W", "A", "W", "A", "R", "W", "W", "R", "A", "G"]
```

Calculate the frequency of alanine (`A`), arginine (`R`), tryptophan (`W`), and glycine (`G`) in this sequence.

````python
amino_acids = ["A", "R", "A", "W", "W", "A", "W", "A", "R", "W", "W", "R", "A", "G"]

# Votre code
````

### <font color="red"> Solution </font>

````python
# if we forbid the use of the .count() method:
amino_acids = ["A", "R", "A", "W", "W", "A", "W", "A", "R", "W", "W", "R", "A", "G"]

count_a = 0
count_r = 0
count_w = 0
count_g = 0

for acid in amino_acids:
    if acid == "A":
        count_a += 1
    elif acid == "R":
        count_r += 1
    elif acid == "W":
        count_w += 1
    elif acid == "G":
        count_g += 1

print("Frequency of alanine:", count_a / len(amino_acids))
print("Frequency of arginine:", count_r / len(amino_acids))
print("Frequency of tryptophan:", count_w / len(amino_acids))
print("Frequency of glycine:", count_g / len(amino_acids))
````

````python
# if we allow the use of the .count() method:
print(
    "Frequency of alanine:",
    amino_acids.count("A") / len(amino_acids),
)
print(
    "Frequency of arginine:",
    amino_acids.count("R") / len(amino_acids),
)
print(
    "Frequency of tryptophan:",
    amino_acids.count("W") / len(amino_acids),
)
print(
    "Frequency of glycine:",
    amino_acids.count("G") / len(amino_acids),
)
````

````python
# a version with list comprehensions:
print(
    "Frequency of alanine:",
    len([acid for acid in amino_acids if acid=="A"]) / len(amino_acids),
)
print(
    "Frequency of alanine:",
    len([acid for acid in amino_acids if acid=="R"]) / len(amino_acids),
)
print(
    "Frequency of alanine:",
    len([acid for acid in amino_acids if acid=="W"]) / len(amino_acids),
)
print(
    "Frequency of alanine:",
    len([acid for acid in amino_acids if acid=="G"]) / len(amino_acids),
)
````

Here are a student's grades: 14, 9, 13, 15, and 12. Create a script that displays the maximum grade (using [`max`](https://docs.python.org/3/library/functions.html#max)), the minimum grade (using [`min`](https://docs.python.org/3/library/functions.html#min)), and calculates the average.

Display the average with two decimal places. Also display the mention obtained, knowing that the mention is `passable` if the average is between 10 inclusive and 12 exclusive, `fairly good` between 12 inclusive and 14 exclusive, and `good` above 14.

````python
# Votre code
````

### <font color="red"> Solution </font>

````python
notes = [14, 9, 13, 15, 12]

moyenne = sum(notes) / len(notes)

if 10 <= moyenne < 12:
    mention = "passable"
elif 12 <= moyenne < 14:
    mention = "assez bien"
elif 14 <= moyenne:
    mention = "bien"
else:
    mention = "aucune"

print("Minimum :", min(notes))
print("Maximum :", max(notes))
print("Moyenne : {:.2f}".format(moyenne))
print("Mention :", mention)
````

## <font color="green">7- Collatz conjecture</font>

The [Collatz (or Syracuse) conjecture](https://en.wikipedia.org/wiki/Collatz_conjecture) is an unsolved mathematical conjecture defined as follows:

Let $n$ be a positive integer. If $n$ is even, divide it by 2. If it is odd, multiply it by 3 and add 1. Repeating this procedure, the sequence reaches the value 1 and then continues indefinitely with a trivial three-value cycle.

So far, the Collatz conjecture, which states that starting from any positive integer the Collatz sequence reaches 1, has never been disproven.

For example, the first elements of the Collatz sequence starting from 10 are: 10, 5, 16, 8, 4, 2, 1...

Create a script that, starting from a positive integer $n$ (for example 10 or 20), creates a list of the Collatz sequence numbers. With different starting points (different values of $n$), is the Syracuse conjecture always verified? What are the numbers that make up the trivial cycle?

Notes:
1. For this exercise, you need to perform an unknown number of iterations until the Collatz sequence reaches 1 and starts its trivial cycle. You can test your algorithm with an arbitrary number of iterations, typically 20 or 100, depending on your starting $n$.
2. A number is even when the remainder of its integer division (operator `%`) by 2 is zero.

````python
# Votre code
````

### <font color="red"> Solution </font>

````python
n = 20
max_iterations = 1000

ns = [n]
iterations = 0
while n != 1:
    if iterations == max_iterations:
        print(f"n = {n} : no convergence in {max_iterations} iterations")
        break
    if n % 2 == 0:
        n //= 2
    else:
        n = n * 3 + 1
    ns.append(n)
    iterations += 1
else:
    print(f"n = {n} : convergence in {iterations} iterations")
    print(ns)
````

## <font color="green">8- Determining prime numbers below 100</font>

Here is an excerpt from [the article on prime numbers](http://fr.wikipedia.org/wiki/Nombre_premier) from Wikipedia.

A prime number is a natural number that has exactly two distinct positive integer divisors (which are 1 and itself). This definition excludes 1, which has only one positive integer divisor. Conversely, a nonzero product of two integers different from 1 is called composite. For example, $6 = 2 \times 3$ is composite, as is $21 = 3 \times 7$, but $11$ is prime because 1 and 11 are the only divisors of 11. The numbers 0 and 1 are neither prime nor composite.

Determine the prime numbers less than 100. How many prime numbers are there between 0 and 100? Here are some suggested methods.

#### Method 1 (less optimal but intuitive)

For each number from 2 to 100, calculate the remainder of integer division (using the modulo `%`) with all numbers from 2 to itself minus 1. If it is a prime number, there will be no number giving a remainder of 0.

#### Method 2 (more optimal and faster but slightly more complex)

Go through all numbers from 2 to 100 and check if they are composite, i.e., they are the product of two prime numbers. Practically, this means checking if the remainder of division (using `%`) between the number considered and each previously determined prime number is zero. If so, the number is not prime. For this method, initialize the list of prime numbers with the first prime number (2!).

````python
# Votre code
````

### <font color="red">Solution</font>

````python
# Method 1
premiers = []
for n in range(2, 101):
    premier = True
    for m in range(2, n):
        if n % m == 0:
            premier = False
            break
    if premier:
        premiers.append(n)
print(premiers)

# Method 1 with for ... else ...
premiers = []
for n in range(2, 101):
    for m in range(2, n):
        if n % m == 0:
            break
    else:
        premiers.append(n)
print(premiers)

# Method 2
premiers = []
for n in range(2, 101):
    for premier in premiers:
        if n % premier == 0:
            break
    else:
        premiers.append(n)
print(premiers)
````

## <font color="green">9- Number search by dichotomy</font>

Binary search ([dichotomy](https://fr.wikipedia.org/wiki/Dichotomy)) is a method that divides a problem (usually in equal parts) to find the solution. For example, here is a discussion between Pierre and Patrick in which Pierre tries to guess the number (between 1 and 100) Patrick is thinking of.

- [Patrick] “I thought of a number between 1 and 100.”
- [Peter] “OK, I will try to guess it. Is your number smaller or larger than 50?”
- [Patrick] “Larger.”
- [Peter] “Is your number smaller, larger, or equal to 75?”
- [Patrick] “Larger.”
- [Peter] “Is your number smaller, larger, or equal to 87?”
- [Patrick] “Smaller.”
- [Peter] “Is your number smaller, larger, or equal to 81?”
- [Patrick] “Smaller.”
- [Peter] “Is your number smaller, larger, or equal to 78?”
- [Patrick] “Larger.”
- [Peter] “Is your number smaller, larger, or equal to 79?”
- [Patrick] “Equal. That’s the number I thought of. Congrats!”

To quickly guess the number, the trick is to take the midpoint of the interval containing the number at each step. Steps:

- Number is between 1 and 100, propose 50 (100 / 2).
- Number is between 50 and 100, propose 75 (50 + (100 - 50)/2).
- Number is between 75 and 100, propose 87 (75 + (100 - 75)/2).
- Number is between 75 and 87, propose 81 (75 + (87 - 75)/2).
- Number is between 75 and 81, propose 78 (75 + (81 - 75)/2).
- Number is between 78 and 81, propose 79 (78 + (81 - 78)/2).

Create a script that reproduces this guessing game. You think of a number between 1 and 100 and the computer tries to guess it by dichotomy, asking you questions.

Your program will use [`input`](https://docs.python.org/3/library/functions.html#input) to interact with the user. Example:

```python

>>> print(letter)

```

For guidance, here's the conversation using the previous example:

```default
Think of a number between 1 and 100.
Is your number larger, smaller, or equal to 50? [+/-/=] +
Is your number larger, smaller, or equal to 75? [+/-/=] +
Is your number larger, smaller, or equal to 87? [+/-/=] -
Is your number larger, smaller, or equal to 81? [+/-/=] -
Is your number larger, smaller, or equal to 78? [+/-/=] +
Is your number larger, smaller, or equal to 79? [+/-/=] =
I found it in 6 questions!
```

The characters `[+/-/=]` indicate to the user how to interact: enter `+` if the chosen number is greater than the computer's guess, `-` if smaller, or `=` if it is equal to the guess (then press Enter).

````python
# Votre code ici
````

### <font color="red">Solution</font>

````python
print("Pensez à un nombre entre 1 et 100.")

bas = 1
haut = 100
iterations = 1
while True:
    moitie = bas + (haut - bas) // 2
    indice = input("Est-ce que votre nombre est plus grand, plus petit ou égal "
                   f"à {moitie} ? [+/-/=] ")
    if indice == "+":
        bas = moitie
    elif indice == "-":
        haut = moitie
    elif indice == "=":
        break
    iterations += 1

print(f"J'ai trouvé en {iterations} questions !")
````
